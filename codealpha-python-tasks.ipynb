# Task 1 HangMan_Game

```python
import random   #To pick a random word

# List of words
words = ["apple", "chair", "table", "rose", "stick"]

# Choose a random word
secret_word = random.choice(words)

# Create blanks
guessed = ["__"] * len(secret_word)

# Number of lives
lives = 6

print("Welcome to Hangman!")
print("Guess the word one letter at a time.\n")

while lives > 0:
    print("Word:", " ".join(guessed))
    print("Lives left:", lives)

    guess = input("Enter a letter: ").lower()

    if len(guess) != 1 or not guess.isalpha():
        print("Please type only ONE letter.\n")
        continue

    # Correct letter
    if guess in secret_word:
        print("Correct!\n")
        for i in range(len(secret_word)):
            if secret_word[i] == guess:
                guessed[i] = guess
    else:
        lives -= 1
        print("Wrong guess!\n")

    # Check win
    if "__" not in guessed:
        print("You wonðŸŽ‰! The word was:", secret_word)
        break

# Game over
if lives == 0:
    print("Game OverðŸ˜¢! The word was:", secret_word)
```

# Summary Of Hnagman Game
Hangman is a classic word-guessing game where the player tries to guess a hidden word one letter at a time. In this project, I created a simple text-based Hangman game using Python. The computer randomly selects a word from a small list of five predefined words. The player is given six chances (lives) to guess the letters correctly.

This program uses basic Python concepts such as lists, strings, loops, if-else conditions, random word selection, and user input. The goal of this task is to practice logical thinking and understand how game loops and conditions work in Python.

# Task 2 Stock Portfolio Tracker

```python
# Stock prices (already fixed)
stocks = {
    "AAPL": 180,
    "TSLA": 250
}

total = 0   # to store total investment

print("Stock Portfolio Tracker")

# take stock name from user
name = input("Enter stock name: ").upper()

# take quantity
qty = int(input("Enter quantity: "))

# calculate investment
price = stocks[name]
investment = price * qty

# add to total 
total = investment

# show result
print("Investment value:", investment)
print("Total investment:", total)
```
